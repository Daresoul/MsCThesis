```python
def update_tokens(row):
    row['toks'] = ['[CLS]'] + row['toks'][0:62] + ['[SEP]']
    return row


dataset = dataset.map(update_tokens)
```

```python
def update_token_to_label(row):
    row['token_to_label'] = [(t, l) for t, l in zip(row['toks'], row['predicted_labels'])]
    return row
        
dataset = dataset.map(update_token_to_label)
```

```python
token_to_label = dataset['token_to_label']
```

```python
from collections import defaultdict

adjectives_seperated_math = defaultdict(int)
adjectives_seperated_display = defaultdict(int)
adjectives_seperated_italic = defaultdict(int)

for token_label_arr in token_to_label:
    for i, (token, label) in enumerate(token_label_arr):
        
        if token == '[MATH]' and label == 'NOUN':
            before = token_label_arr[max(i-3, 0):i]
            for t, label in before:
                if label in ['ADJ']:
                    adjectives_seperated_math[t] += 1
                    
        if token == '[DISPLAY]' and label == 'NOUN':
            before = token_label_arr[max(i-3, 0):i]
            for t, label in before:
                if label in ['ADJ']:
                    adjectives_seperated_display[t] += 1
                    
        if token == '[ITALIC]' and label == 'NOUN':
            before = token_label_arr[max(i-3, 0):i]
            for t, label in before:
                if label in ['ADJ']:
                    adjectives_seperated_italic[t] += 1
```

```python
def add_together(arr, arr2):
    for t, value in arr2.items():
        arr[t] += value

x = defaultdict(int)

add_together(x, adjectives_seperated_math)
add_together(x,adjectives_seperated_display)
add_together(x,adjectives_seperated_italic)
```

```python
import random

def get_token_as_label(token, label, num=1, max_len=200):
    arr = []
    
    for token_labels in token_to_label:
        if len(token_labels) > max_len:
            continue
        for t, l in token_labels:
            if t == token and l == label:
                arr.append(token_labels)
        
    random_items = random.sample(arr, num)
    
    return random_items

def arrays_to_text(arr):
    for sentence in arr:
        for token, label in sentence:
            print("("+token+", "+label+") ", end='')
        print()
    
        
```
